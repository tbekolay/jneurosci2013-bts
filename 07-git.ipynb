## Let Github host supplementary material

Ideally, you'd release the source
for your research and your paper as supplementary material.
Since having the history can help
interested readers dig deeper,
version control is great way to do that.
But, it's possible that you'll later
amend parts of your source code.
One reviewer wrote:

> Will the Nengo scripts be available through J Neuroscience? I'm not
> quite sure what the policy is for this kind of material but it would
> be nice if it were available with the article as the github may change over time ...

To address this, I tagged the commit
that contained the files used to generate
the actual submission.

```python
!git tag
```

```python
!git show --quiet 2013-11-29
```

Github has a nice feature called **Releases**
that is tied to `git`'s tagging system.
Once you push a tag to Github,
you can add some information to it and even
attach arbitrary files.

[2013-11-29 tag](https://github.com/tbekolay/jneurosci2013/releases/tag/2013-11-29)
