# Exploring with the IPython notebook

The notebook has matured recently,
and is now an incredibly powerful tool
for serious analysis.
I primarily use it for exploratory work
as it's very easy to iterate.

```python
import numpy as np
np.random?
```

```python
# Analysis block 1
data = np.random.randn(200)
std = np.std(data)
mean = np.mean(data)
```

```python
# Plotting block 1
import matplotlib.pyplot as plt
plt.plot(data)
plt.axhline(mean, color='k')
plt.fill_between(range(data.size), mean-std, mean+std, alpha=0.3);
```

```python
# Analysis block 2
def normalize(x):
    return x / np.max(x)
n_data = normalize(data)
```

```python
# Plotting block 2
import matplotlib.pyplot as plt
plt.plot(n_data)
```

# Autoreload extensions

Something I will commonly do is
edit some scripts in a text editor
while also playing around with it
in an IPython notebook.
To avoid constantly restarting the notebook,
you can instead set up IPython
to automatically reload extensions.

```python
!cat test.py
```

```python
import test
test.do_thing()
```

```python
!cat test.py
```

```python
test.do_thing()
```

```python
%load_ext autoreload
```

```python
autoreload?
```

```python
autoreload 2
```

```python
test.do_thing()
```

```python
test.do_thing()
```
