# Combining plots into figures

A typical figure is not just one plot.
Most people make their plots with scripts,
but then import those plots into
Adobe Illustrator or Inkscape to combine them.
This makes it difficult to change the underlying plot later,
makes figures harder to reproduce,
and creates the temptation to tinker with plots after the fact.

`svgutil` is a small Python package that can
combine several SVGs into a single SVG,
which can then be converted to many other formats.

```python
import neo
import matplotlib.pyplot as plt
import numpy as np
import quantities as pq
plexon = neo.io.PlexonIO(filename='data.plx')
data = plexon.read(cascade=True, lazy=False)[0]
seg = data.segments[0]
plt.figure(figsize=(8,4))
plt.title(seg.file_origin)
trains = [st.rescale('s').magnitude for st in seg.spiketrains]
colors = plt.cm.jet(np.linspace(0, 1, len(seg.spiketrains)))
plt.ylim(len(trains)-0.5, -0.5)
plt.xlim(0, 5)
plt.ylabel('Neuron')
plt.xlabel('Time (s)')
plt.eventplot(trains, colors=colors, rasterized=True)
plt.savefig('plexon.svg')
plt.close()

from IPython.display import SVG
SVG(filename='plexon.svg')
```

```python
import neo
import matplotlib.pyplot as plt
import numpy as np
import quantities as pq
nex = neo.io.NeuroExplorerIO(filename='data.nex')
data = nex.read(cascade=True, lazy=False)[0]
seg = data.segments[0]
plt.figure(figsize=(8,4))
plt.title(seg.file_origin)
trains = [st.rescale('s').magnitude for st in seg.spiketrains]
colors = plt.cm.jet(np.linspace(0, 1, len(seg.spiketrains)))
plt.ylim(len(trains)-0.5, -0.5)
plt.xlim(0, 5)
plt.ylabel('Neuron')
plt.xlabel('Time (s)')
plt.eventplot(trains, colors=colors, rasterized=True)
plt.savefig('nex.svg')
plt.close()

from IPython.display import SVG
SVG(filename='nex.svg')
```

```python
import svgutils.transform as sg

# Create new SVG figure
fig = sg.SVGFigure(str(8 * 72), str(8 * 72))

# Load matpotlib-generated figures
fig1 = sg.fromfile('plexon.svg')
fig2 = sg.fromfile('nex.svg')

# Get the plot objects
plot1 = fig1.getroot()
plot2 = fig2.getroot()
plot2.moveto(0, 4 * 72)

# Add text labels
txt1 = sg.TextElement(25, 20, "A", size=24, weight="bold")
txt2 = sg.TextElement(25, 20 + 4 * 72, "B", size=24, weight="bold")

# Append plots and labels to figure
fig.append([plot1, plot2])
fig.append([txt1, txt2])

# Save generated SVG files
fig.save("figure.svg")

from IPython.display import SVG
SVG(filename='figure.svg')
```
